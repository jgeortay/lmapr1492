```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.analysis.diffraction.xrd import *
from matplotlib import pyplot as plt
from IPython.display import Markdown,display
import numpy as np
def printbold(string): 
    display(Markdown(string))
    
printbold("**Téléchargement du fichier CIF...**")

with MPRester("gsItyjn65HPAFPdGywpu") as m:
    structure = m.get_structure_by_material_id("mp-47")
    w = CifWriter(structure, 0.1)
    w.write_file('C_mp-47_symmetrized.cif')
    
printbold("**Diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å :**")

xrd = XRDCalculator()
xrd_pattern = xrd.get_pattern(structure)
xrd.get_plot(structure,annotate_peaks=True)
plt.show()
print("Sur le diffractogramme les trois premiers pics correspondent à des angles compris entre 40 et 50 degrés.")
hkls = xrd_pattern.hkls
print("Les indices hkl correpondants sont :")
print("        - Pour le premier pic, le programme nous renvoie :",hkls[0])
print("          Plus simplement : hkl = (1, 0, -1)")
print("        - Pour le deuxième pic, le programme nous renvoie :",hkls[1])
print("          Plus simplement : hkl = (0, 0, 0)")
print("        - Pour le troisième pic, le programme nous renvoie :",hkls[2])
print("          Plus simplement : hkl = (1, 0, -1)")
```
